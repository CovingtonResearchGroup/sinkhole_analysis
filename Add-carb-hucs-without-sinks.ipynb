```python
hucs_on_carbs = gpd.read_file('huc12_on_carbs.shp')
processed_hucs = gpd.read_file('carb_huc_dems/processed_hucs.shp')
want_crs = processed_hucs.crs 

```

```python
hucs_on_carbs = hucs_on_carbs.to_crs(want_crs)

hucs_on_carbs.crs
```

```python
zero_sink_hucs = []
for i, row in hucs_on_carbs.iterrows():
    if len(processed_hucs[processed_hucs.huc12 == row.huc12]) == 0:
        zero_sink_hucs.append(row)
```

```python
df = gpd.GeoDataFrame(zero_sink_hucs, crs = want_crs)
df
```

```python
df.crs
```

```python
df_no_big = df[df.areasqkm<1000]
```

```python
df_no_big.plot()
```

```python
from sinkhole_functions import get_carbs_only_huc
```

```python
df_trimmed = df_no_big
for idx, row in df_no_big.iterrows():
    if idx % 100 == 0:
        print(idx)
    geom = row.geometry
    huc_mask = gpd.GeoSeries([geom], crs=want_crs)
    trimmed_huc = get_carbs_only_huc(huc_mask)
    df_trimmed.at[idx, 'geometry'] = trimmed_huc
```
